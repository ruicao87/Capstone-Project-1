```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import nltk
%matplotlib inline
from collections import Counter
import nltk
from nltk.corpus import stopwords 
from textblob import TextBlob
from nltk.util import ngrams
from nltk.tokenize import WhitespaceTokenizer
from nltk.stem import WordNetLemmatizer
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
```

```python
df = pd.read_csv('Womens Clothing E-Commerce Reviews.csv')
```

```python
df.head()
```

```python
df = df[df['Review Text'].notnull()]
```

```python
df['Review Length']= df['Review Text'].apply(len)
```

```python
df['review text']=df['Review Text'].str.lower()
```

```python
df['review text'] = df['review text'].str.strip()
```

```python
df['review text'] = df['review text'].str.replace('\d+','')
```

```python
df['review text'] = df['review text'].str.replace('[^\w\s]', '')
```

```python
tokenizer = WhitespaceTokenizer()
lemmatizer = WordNetLemmatizer()
def lemmatize_text(text):
      return [lemmatizer.lemmatize(w) for w in tokenizer.tokenize(text)]
# Apply the word lemmatizer function to data
df['review text'] = df['review text'].astype(str).apply(lemmatize_text)
```

```python
stop_words = set(stopwords.words('english'))
def remove_stopwords(text):
    return[w for w in text if not w in stop_words]
df['review text']= df['review text'].apply(remove_stopwords)
```

```python
df.head()
```

```python
count = CountVectorizer(min_df=.005,ngram_range=(1,2))
bag_of_words = count.fit_transform(df['review text'].astype(str))
len(count.vocabulary_)
```

```python
feature_names = count.get_feature_names()

# View feature names
feature_names
```

```python
words_count = pd.DataFrame(bag_of_words.toarray(),columns =feature_names )
words_count.head(10)
```

```python
vector = TfidfVectorizer(min_df=.005, ngram_range=(2,3))
vector_frequency = vector.fit_transform(df['review text'].astype(str))
len(vector.vocabulary_)
```

```python
words_frequency = pd.DataFrame(vector_frequency.toarray(),columns = vector.get_feature_names())
words_frequency.head(10)
```
